```python
import numpy as np
import pandas as pd
import time
from mlxtend.frequent_patterns import apriori
```

```python
df = pd.read_csv('/content/Chess_data.csv',sep='\s+',header=None)

trans=[]
for ind in df.index:
  trans.append([int(i) for i in df[0][ind].split()])
trans
```

```python
unique_items = set()
for ticker in df:
    unique_items.update(df[ticker].unique())
unique_items={x for x in unique_items if x==x}
print(unique_items)
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
start = time.time()
frequent_items = apriori(df, min_support = 0.5, use_colnames = True)
end = time.time()
print(frequent_items)
```

```python
time_taken=end-start
time_taken
```
