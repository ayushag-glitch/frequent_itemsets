```python
import numpy as np
import pandas as pd
import time
from mlxtend.frequent_patterns import apriori
df = pd.read_csv('/content/liquor.csv',sep='\s+',header=None)

trans=[]
for ind in df.index:
  trans.append([int(i) for i in df[0][ind].split()])
trans
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
start = time.time()
frequent_items = apriori(df, min_support = 0.02, use_colnames = True)
end = time.time()
print(frequent_items)
```

```python
time_taken=end-start
time_taken
```
