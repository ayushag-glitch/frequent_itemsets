```python
import time
import pandas as pd
df = pd.read_csv('Chess_data.csv',sep='\s+',header=None)
# print(df)
trans=[]
for ind in df.index:
  trans.append([int(i) for i in df[0][ind].split()])
trans
```

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
start = time.time()
print(fpgrowth(df, min_support=0.5, use_colnames=True))
end=time.time()
```

```python
time_taken=end-start
time_taken
```
