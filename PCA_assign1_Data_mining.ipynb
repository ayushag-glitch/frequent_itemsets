```python
import pandas as pd
import numpy as np
```

```python
# UPLOAD THE DATA FILE TO GOOGLE COLAB TO RUN THE FOLLOWING CODE, ELSE EDIT THE PATH IN THE LINE BELOW
origdf = pd.read_csv('/content/movement_libras.data', header=None)
df=origdf
# Naming the columns
column_names=["Feature"+str(i) for i in range(1,91)]
column_names.append("Class_num")
df.columns = column_names
# Printing dataframe
print(df)
```

```python
class_list=df['Class_num']
# Dropped the Class number column since that is not a feature
df.drop('Class_num', axis=1, inplace=True)
df
```

```python
column_means = {}
column_standard_devs={}
# Calculating and printing mean and standard deviations of the columns
for ticker in df.columns:
  column_means[ticker]=df[ticker].mean()
  column_standard_devs[ticker]=df[ticker].std()
print(column_means)
print(column_standard_devs)
```

```python
# Finding covariance matrix
for ticker in df.columns:
  df[ticker]=(df[ticker]-df[ticker].mean())/df[ticker].std()
df=df.cov()
# Calculating eigen vectors and values
eigval, eigvec=np.linalg.eig(df)
# Sorting eig values and vectors
idx = eigval.argsort()[::-1]   
eigval = eigval[idx]
eigvec = eigvec[:,idx]
# mapping eigenvalues to eigenvectors in a dictionary and printing
eig = {eigval[i]: eigvec[i] for i in range(len(eigval))}
for ticker in eig:
  print("For eigval ", ticker, "Eig vec is:", eig[ticker])
# Calculating the percentage of variation with each eigenvector
eigvalsum=eigval.sum()
perc_variance=[]
perc_variance_sum=[]
perc_sum=0
pc_count=0
# CHANGE THE BELOW VALUE TO GET MINIMUM DESIRED VARIANCE COVERAGE WITH THE PCA
# For values greater than 100, 100 is the value considered as should be
desired_variance_coverage=99
for ticker in eigval:
  temp=ticker*100/eigvalsum
  perc_sum+=temp
  perc_variance.append(temp)
  perc_variance_sum.append(perc_sum)
for ticker in perc_variance_sum:
  pc_count+=1
  if ticker>desired_variance_coverage:
    break
```

```python
# Printing the arrays containing the percentage variance, aggregrate percentage variance till that index,
# and the number of principal components for the desired variance coverage
print(perc_variance)
print(perc_variance_sum)
print(pc_count)
```

```python
# Calculating the tranformed data using the required eigenvectors and printing
eigvec=eigvec[:pc_count]
eigvec=eigvec.transpose()
transformed_data=pd.DataFrame(np.matmul(np.array(origdf),eigvec), columns=["PC_"+str(i) for i in range(1,pc_count+1)])
transformed_data
```
